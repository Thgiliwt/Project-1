# Part 1b Attendance rate per school type
This notebook reads the school attendance rate data and generates a bar graph of attendance rate for each school type.  

## Output  
Writes the bar graph to ```../data/fig2_part1.png```

````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
````

````python
# read in the data and list the first few lines to confirm that all is ok
attendance_metadata = "..\data\cleaned_data_schoolattendancerate_seifa.csv"
attendance_df = pd.read_csv(attendance_metadata)
attendance_df.head(10)
````

````python
# calculate the mean attendance rate for each school type for 2015 - 2019 
for year in range(2015,2020):
    attendance_year_df=attendance_df.loc[attendance_df['Attendance Year']==year]
    mean_year=attendance_year_df.groupby(['School Type'])['Attendance Rate'].mean()
    print(year, mean_year)
````

````python
# drop the columns that I do not need for the analysis and pivot on Attendance Year 
df2 =attendance_df.iloc[:,[2,3,11]]
df3 = df2.groupby(['Attendance Year','School Type']).mean().reset_index()
df = df3.pivot(index='Attendance Year', columns='School Type',values='Attendance Rate')
df
````

````python
# make the bar graph and write it out 
a=df.plot(kind="bar",title="Attendance Rate VS School Type",figsize=(12,8)).legend(
    loc='upper center', ncol=3, title="School Type")
plt.ylabel('Attendance Rate')
plt.savefig('../data/fig2_part1.png')
plt.show()
````
